# **Reconstruction via le théorème de la tranche de Fourier**
## ***Question 5***

```python
def reconFourierSlice(sinogram):
    N = sinogram.shape[0]
    res = np.zeros((N,N))
    for i in range(N):
        res[i] = np.fft.ifft(sinogram[i]).real
    return res

```
